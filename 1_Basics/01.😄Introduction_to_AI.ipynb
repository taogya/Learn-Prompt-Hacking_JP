# AI とは何か？

**最終更新**: `2024-08-12`

これは非常に簡略化された定義です。詳細は [Wikipedia](https://en.wikipedia.org/wiki/Artificial_intelligence) を参照してください ↩

人工知能（AI）は、データから「学習」できる「賢い」アルゴリズムを作成する分野です。

本コースでは **生成 AI（GenAI）** に焦点を当てます。生成 AI とは、これまで人間にしかできなかった記事や画像などの新しいコンテンツを作成できる AI の一種です。例えば、生成 AI はエッセイを書いたり、数学の問題を解いたり、アートコンテストで優勝したりできます。

生成 AI は非常に便利ですが、何をすべきかを人間が指示する必要があります。多くの場合、生成 AI は新入社員のようなものです — 非常に有能ですが、良い仕事をするためには明確な指示が必要です。生成 AI を適切に指示できることは非常に強力なスキルです。生成 AI には **プロンプト**（通常はテキストによる指示、例：「アイスクリームのフレーバーを5つ挙げて」）を送ることで指示します。テキストを読み書きする生成 AI は **大規模言語モデル（LLM）** と呼ばれます。

**プロンプトはどの言語でも書くことができます。また、後のコースで見るように、プロンプトは画像、音声、その他のメディアで構成されることもあります。**
